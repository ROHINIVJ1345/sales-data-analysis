```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#load the dataset

sales_data=pd.read_csv("/content/Sales_Data.csv")

```

```python
#remove unwanted columns

sales_data.drop(['Unnamed: 0','Purchase Address'],axis=1,inplace=True)
```

```python
sales_data.head()
```

```python
#duplicate

sales_data.duplicated().sum()


```

```python
#missing values

sales_data.isnull().sum()
```

```python
#extract

sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data['month'] = sales_data['Order Date'].dt.month
sales_data['year'] = sales_data['Order Date'].dt.year
sales_data['Product Category'] = sales_data['Product'].str.split().str[0]
sales_data['Sales'] = pd.to_numeric(sales_data['Sales'], errors='coerce')

```

```python
sales_data.head()
```

```python
sales_data.describe()
```

```python


# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numeric column
Q1 = sales_data.select_dtypes(include=np.number).quantile(0.25)
Q3 = sales_data.select_dtypes(include=np.number).quantile(0.75)

# Calculate the IQR for each numeric column
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify rows that have outliers
is_outlier = (sales_data.select_dtypes(include=np.number) < lower_bound) | (sales_data.select_dtypes(include=np.number) > upper_bound)

# Show the rows that contain outliers
outliers = sales_data[is_outlier.any(axis=1)]
print("Outliers found:")
print(outliers)
```

```python
# Creating a box plot for the 'Sales' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=sales_data['Sales'], color='skyblue')
plt.title('Box Plot of Sales Data')
plt.xlabel('Sales')
plt.show()
```

```python
# Use Seaborn or Matplotlib to create visualizations
#LINE PLOT
# Convert 'date' to datetime format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])

```

```python
#aggregate sales over time
sales_over_time = sales_data.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
# Plot sales over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=sales_over_time, x='Order Date', y='Sales')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate sales by product category
sales_product_category = sales_data.groupby('Product Category')['Sales'].sum().reset_index()

# Bar plot for sales by Product_category
plt.figure(figsize=(10, 5))
sns.barplot(data=sales_product_category, x='Product Category', y='Sales',color= '#185519')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
3.# ADVANCE ANALYSIS

# Convert 'Order Date' to datetime format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])

# Set 'Order Date' as index
sales_data = sales_data.set_index('Order Date')

# Calculate monthly sales
monthly_sales = sales_data['Sales'].resample('M').sum()

# Calculate percentage change month-over-month
monthly_sales_pct_change = monthly_sales.pct_change() * 100

# Plotting the sales and percentage change
plt.figure(figsize=(14, 6))

# Plot total sales
plt.subplot(2, 1, 1)
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales')
plt.ylabel('Total Sales')
plt.grid(True)

# Plot percentage change
plt.subplot(2, 1, 2)
plt.plot(monthly_sales_pct_change, marker='o', linestyle='-', color='orange')
plt.title('Percentage Change in Sales Month-over-Month')
plt.ylabel('Percentage Change (%)')
plt.grid(True)

plt.tight_layout()
plt.show()

# Analyzing Seasonality (requires data for multiple years)
# For simple seasonality, let's plot sales by month for each year
sales_data['Month'] = sales_data.index.month
sales_data['Year'] = sales_data.index.year
```
